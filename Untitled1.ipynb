```python
from google.colab import files
import pandas as pd

uploaded = files.upload()
csv_name = list(uploaded.keys())[0]

df = pd.read_csv(csv_name)
print("Loaded:", csv_name, "Shape:", df.shape)
df.head()
```

```python
import numpy as np

print("Initial shape:", df.shape)

# Drop exact duplicates
df = df.drop_duplicates()
print("After duplicates removed:", df.shape)

# Basic missing values check
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values (top):")
print(missing.head(10))

# Fix/convert album release date -> year (if column exists)
if "track_album_release_date" in df.columns:
    df["track_album_release_date"] = pd.to_datetime(df["track_album_release_date"], errors="coerce")
    df["release_year"] = df["track_album_release_date"].dt.year

# Optional: fill missing numeric values with median
num_cols = df.select_dtypes(include=np.number).columns
for c in num_cols:
    df[c] = df[c].fillna(df[c].median())

# Optional: fill missing text with "Unknown"
obj_cols = df.select_dtypes(exclude=np.number).columns
for c in obj_cols:
    df[c] = df[c].fillna("Unknown")

print("Done preprocessing. Shape:", df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt

# Summary
display(df.describe(include="all").T.head(20))

# Popularity distribution
plt.figure()
df["track_popularity"].hist(bins=30)
plt.title("Track Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Count")
plt.show()

# Genres count (top 15)
plt.figure()
df["playlist_genre"].value_counts().head(15).plot(kind="bar")
plt.title("Top Playlist Genres (Count)")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()

# Tempo vs Energy scatter (sample for speed)
sample = df.sample(3000, random_state=42)
plt.figure()
plt.scatter(sample["tempo"], sample["energy"])
plt.title("Tempo vs Energy (Sample)")
plt.xlabel("Tempo")
plt.ylabel("Energy")
plt.show()

# Boxplot-like view using quantiles (quick insight)
features_quick = ["danceability","energy","speechiness","acousticness","instrumentalness","liveness","valence","tempo","loudness"]
display(df[features_quick].describe(percentiles=[.05,.25,.5,.75,.95]).T)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Audio features for correlation
audio_features = [
    "danceability","energy","key","loudness","mode","speechiness",
    "acousticness","instrumentalness","liveness","valence","tempo","duration_ms",
    "track_popularity"
]

use_features = [c for c in audio_features if c in df.columns]
corr = df[use_features].corr(numeric_only=True)

plt.figure(figsize=(10, 8))
plt.imshow(corr, aspect="auto")
plt.title("Correlation Matrix (Features)")
plt.xticks(range(len(use_features)), use_features, rotation=90)
plt.yticks(range(len(use_features)), use_features)
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X = df[use_features].copy()

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow method
inertias = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

plt.figure()
plt.plot(list(K_range), inertias, marker='o')
plt.title("Elbow Method (KMeans)")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()

# Silhouette scores
sil_scores = []
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    sil = silhouette_score(X_scaled, labels)
    sil_scores.append(sil)

plt.figure()
plt.plot(list(K_range), sil_scores, marker='o')
plt.title("Silhouette Score vs K")
plt.xlabel("K")
plt.ylabel("Silhouette Score")
plt.show()

best_k = list(K_range)[int(np.argmax(sil_scores))]
print("Suggested best K (by silhouette):", best_k)
```

```python
from sklearn.decomposition import PCA

k = best_k  # or set manually like k=6
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df["cluster"] = kmeans.fit_predict(X_scaled)

print(df["cluster"].value_counts().sort_index())

# PCA for 2D visualization
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

plt.figure()
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df["cluster"], s=8)
plt.title("Clusters Visualized using PCA (2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Cluster counts per genre
genre_cluster = pd.crosstab(df["playlist_genre"], df["cluster"])
display(genre_cluster)

# Plot (top genres only for readability)
top_genres = df["playlist_genre"].value_counts().head(10).index
genre_cluster_top = genre_cluster.loc[top_genres]

plt.figure(figsize=(10,5))
plt.imshow(genre_cluster_top.values, aspect="auto")
plt.title("Cluster Distribution for Top 10 Genres")
plt.xticks(range(genre_cluster_top.shape[1]), genre_cluster_top.columns)
plt.yticks(range(genre_cluster_top.shape[0]), genre_cluster_top.index)
plt.colorbar()
plt.tight_layout()
plt.show()
print("Best K:", best_k)
print("Silhouette Score:", max(sil_scores))
print("Cluster Distribution:")
print(df["cluster"].value_counts())
print("Top genres:", list(top_genres))
```

```python
top_playlists = df["playlist_name"].value_counts().head(10).index
pl_cluster = pd.crosstab(df[df["playlist_name"].isin(top_playlists)]["playlist_name"],
                        df[df["playlist_name"].isin(top_playlists)]["cluster"])
display(pl_cluster)

plt.figure(figsize=(10,5))
plt.imshow(pl_cluster.values, aspect="auto")
plt.title("Cluster Distribution for Top 10 Playlist Names")
plt.xticks(range(pl_cluster.shape[1]), pl_cluster.columns)
plt.yticks(range(pl_cluster.shape[0]), pl_cluster.index)
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors

# Fit a nearest-neighbor model on scaled features
nn = NearestNeighbors(n_neighbors=15, metric="euclidean")
nn.fit(X_scaled)

def recommend_by_track_name(track_name, n=10):
    # Find matching rows
    matches = df[df["track_name"].str.lower() == track_name.lower()]
    if matches.empty:
        print("Track not found. Try exact name.")
        return

    idx = matches.index[0]
    distances, indices = nn.kneighbors([X_scaled[idx]], n_neighbors=n+1)  # +1 because it returns itself
    rec_idxs = indices[0][1:]  # skip itself

    recs = df.loc[rec_idxs, ["track_name","track_artist","playlist_genre","track_popularity","cluster"]].copy()
    recs["distance"] = distances[0][1:]
    return recs.reset_index(drop=True)

# Example:
recommend_by_track_name("Shape of You", n=10)
```

```python
## Conclusion

Based on the Elbow and Silhouette method, the optimal number of clusters is 2.

The silhouette score of 0.135 indicates moderate cluster separation.

Songs were grouped using audio features like danceability, energy, loudness, tempo, and valence.

Popular genres such as EDM, Rap, Pop, R&B, Latin, and Rock appear across clusters, showing different musical patterns.

Overall, the KMeans clustering model successfully segmented Spotify songs based on audio characteristics.
```
